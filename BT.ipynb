```python
import os
import glob
import pandas as pd


data_folder = r"C:\Users\ANEENA THOMAS\OneDrive - University of Aberdeen\GENES"

file_pattern = os.path.join(data_folder, "*_data_file.txt")


all_data = []

for file_path in glob.glob(file_pattern):
    try:
        df = pd.read_csv(file_path, sep="\t") 
        df['SourceFile'] = os.path.basename(file_path)
        all_data.append(df)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")


if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    print(" Successfully loaded all gene files!")
    display(combined_df)  
else:
    print(" No files found. Check your path or file names.")
```

```python
from difflib import SequenceMatcher


def classify_mutation(wt_seq, mut_seq):
    if wt_seq == mut_seq:
        return "No Change"
    if len(wt_seq) == len(mut_seq):
        return "Substitution"
    elif len(wt_seq) < len(mut_seq):
        return "Insertion"
    else:
        return "Deletion"


def classify_location(wt_seq, mut_seq, promoter_length=1000):
    return "Promoter" if wt_seq[:promoter_length] != mut_seq[:promoter_length] else "Coding"


combined_df["MutationType"] = combined_df.apply(
    lambda row: classify_mutation(row["WildType.Sequence"], row["Mutant.Sequence"]), axis=1)

combined_df["MutationRegion"] = combined_df.apply(
    lambda row: classify_location(row["WildType.Sequence"], row["Mutant.Sequence"]), axis=1)


combined_df[["Gene", "MutationType", "MutationRegion"]]
```

```python
for assay in ['mRNA.Expression', 'Protein.Expression', 'CellViability']:
    for state in ['WT', 'Mut']:
        rep_cols = [f"{assay}.{state}.Rep{i}" for i in (1,2,3)]
        mean_col  = f"{assay}.{state}.Mean"
        combined_df[mean_col] = combined_df[rep_cols].mean(axis=1)
```

```python
print([c for c in combined_df.columns if c.endswith('.Mean')])

combined_df[[
    'mRNA.Expression.WT.Mean','mRNA.Expression.Mut.Mean',
    'Protein.Expression.WT.Mean','Protein.Expression.Mut.Mean',
    'CellViability.WT.Mean','CellViability.Mut.Mean'
]]
```

```python
combined_df['ΔmRNA']      = combined_df['mRNA.Expression.Mut.Mean']  / combined_df['mRNA.Expression.WT.Mean']
combined_df['ΔProtein']   = combined_df['Protein.Expression.Mut.Mean'] / combined_df['Protein.Expression.WT.Mean']
combined_df['ΔViability'] = combined_df['CellViability.Mut.Mean']    / combined_df['CellViability.WT.Mean']

combined_df[['Gene','ΔmRNA','ΔProtein','ΔViability']]
```

```python
combined_df['Score'] = (
      (combined_df['ΔProtein']  - 1)    
    - (combined_df['ΔViability'] - 1)    
    + 0.5*(combined_df['ΔmRNA']     - 1) 
)


top5 = combined_df.sort_values('Score', ascending=False).head(5)


top5[['Gene','MutationType','MutationRegion',
      'ΔmRNA','ΔProtein','ΔViability','Score']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


combined_df['Top5_Label'] = combined_df['Gene'].apply(
    lambda x: 'Top 5' if x in ['Dsgt9a1', 'Ghos8', 'Jok4a3', 'Bma9', 'App1l8'] else 'Other'
)


sorted_df = combined_df.sort_values(by='ΔViability', ascending=False)


plt.figure(figsize=(14, 6))
sns.barplot(
    x='Gene',
    y='ΔViability',
    data=sorted_df,
    hue='Top5_Label',
    dodge=False,
    palette={'Top 5': 'tomato', 'Other': 'skyblue'}
)

plt.xticks(rotation=90)
plt.title("ΔViability of All Genes (Top 5 Highlighted)")
plt.xlabel("Gene")
plt.ylabel("ΔViability")
plt.legend(title="Gene Category")
plt.tight_layout()
plt.show()
```

```python
top5_genes = ['Dsgt9a1', 'Ghos8', 'Jok4a3', 'Bma9', 'App1l8']


combined_df['Top5'] = combined_df['Gene'].apply(lambda x: 'Top 5' if x in top5_genes else 'Other')

plt.figure(figsize=(10, 7))

sns.scatterplot(
    data=combined_df,
    x='ΔmRNA',
    y='ΔProtein',
    hue='MutationType',
    style='Top5',
    s=100,
    palette='Set2',
    edgecolor='black'
)

plt.title("ΔmRNA vs ΔProtein (Top 5 Genes Highlighted)")
plt.legend(title='Mutation Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=combined_df,
    x='MutationRegion',
    y='ΔViability',
    palette='pastel'
)


top5_df = combined_df[combined_df['Gene'].isin(top5_genes)]
sns.stripplot(
    data=top5_df,
    x='MutationRegion',
    y='ΔViability',
    hue='Gene',
    dodge=True,
    jitter=True,
    size=8,
    marker='D',
    edgecolor='black',
    palette='Set1',
    linewidth=1.2
)

plt.title("ΔViability by Mutation Region with Top 5 Genes Overlaid")
plt.legend(title='Top 5 Genes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
